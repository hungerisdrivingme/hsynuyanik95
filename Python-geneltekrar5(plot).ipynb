```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline 
```

```python
yaslistesi=[10,12,14,15,17,22,27]
kilolistesi=[20,35,43,48,57,65,77]
```

```python
npyas=np.array(yaslistesi)
npkilo=np.array(kilolistesi)
```

```python
npyas,npkilo
```

```python
plt.plot(npyas,npkilo,"r")

plt.xlabel("yas")
plt.ylabel("kilo")

plt.title("Grafik")
```

```python
numpydizi=np.linspace(0,10,20)
numpydizi
```

```python
numpydizi2=numpydizi**3
```

```python
plt.plot(numpydizi,numpydizi2,"r*-")
plt.xlabel("normali")
plt.ylabel("kübü")
plt.title("kübün grafigi")
```

```python
plt.subplot(1,2,1)
plt.plot(numpydizi,numpydizi2,"g*-")
plt.subplot(1,2,2)  #1 sıra,2. kolon olacak, 2. grafik
plt.plot(numpydizi2,numpydizi,"b--")
```

```python
benimFigure=plt.figure()
figureaxes=benimFigure.add_axes([0.2,0.2,0.9,0.9])
figureaxes.plot(numpydizi,numpydizi2,"g")
figureaxes.set_xlabel("X ekseni")
figureaxes.set_ylabel("Y ekseni")
figureaxes.set_title("grafikbaslıgı")
```

```python
benimFigure2=plt.figure()
eksen1=benimFigure2.add_axes([0.1,0.1,0.9,0.9])
eksen2=benimFigure2.add_axes([0.6,0.2,0.3,0.3])
eksen1.plot(numpydizi2,numpydizi,"black")
eksen2.plot(numpydizi,numpydizi2,"r")

eksen1.set_xlabel("x")
eksen1.set_ylabel("y")
eksen1.set_title("grap1")
eksen2.set_xlabel("x")
eksen2.set_ylabel("y")
eksen2.set_title("grap2")
```

```python
(benimFigure,benimEksenler)=plt.subplots(nrows=1,ncols=2)

for eksen in benimEksenler:
    eksen.plot(numpydizi,numpydizi2,"b")
    eksen.set_xlabel("x ekseni")
plt.tight_layout()
```

```python
YeniFigur=plt.figure(dpi=100)
Yenieksen=YeniFigur.add_axes([0.1,0.1,0.9,0.9])
Yenieksen.plot(numpydizi,numpydizi**2,"r*-",label="numpydizisi**2")
Yenieksen.plot(numpydizi,numpydizi**3,"b--",label="nmpydizi**3")
Yenieksen.legend(loc=2)
```
